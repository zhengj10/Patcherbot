```python
import sys
import os
import tifffile as tff
import numpy as np
sys.path.append('/Users/zhengj10/Python/Modules')
import PatcherBot as pb
import matplotlib.pyplot as plt
```

```python
imgPath = '/Volumes/Backup Plus/PatcherBot/raw'
datePath = ['2021.05.19_000']
```

```python
fig = plt.figure(figsize=[12,6])
i = 1
for date in datePath:
    platePath = ['plate_000']
    for plate in platePath:
        cellPath = pb.getPathInfo(os.path.join(imgPath, date, plate))
        for cell in(cellPath):
            imgFile = os.path.join(imgPath, date, plate, cell, 'fluor_image_000.tif')
            if not os.path.exists(imgFile):
                continue
            if plate == 'plate_000' and cell in ['patch_attempt_0001_000','patch_attempt_0018_000']:
                continue
            imgFile = tff.TiffFile(imgFile)
            imgFile = imgFile.asarray()
            ax = plt.subplot(2,4,i)
            ax.imshow(imgFile, cmap='gray',vmax=6000)
            i+=1
            ax.axis('off')
plt.tight_layout()
fig.savefig('/Users/zhengj10/Desktop/210526/Voltron 2-ST & JF549/fluor_img_BAPTA_JF549_200nM.png')
plt.show()
```
